Imputation Methods

Variables can be classified into two types numerical and categorical.
For both the imputation process will be different.

For Numerical Variables:
  1. Mean Imputation: Replace all the null values of a column with its mean
     [similar thing can be done with the median]
  2. Arbitrary Value Imputation: Add values far away from the range of column typically very large or very small values.
  3. KNN Imputation: using the similarity between points it could calculate possible value for null.

For Both:  
  1. Frequent Category Imputation: Replace all the null values of a column with its mean().
  2. String Imputation: Replace all values with string like "missing" or "NA"




```python
# I will be using Mean Imputation
#Some pros and cons are as follows:
# Pros:
#   works well with large datasets
#   implementation is simple
#   calculation is simple
# Cons:
#   reduces variance and std deviation
#   could introduce some bias if there are a lot of null values
#   in a skewed distribution could be affected by the outliers

from pandas.api.types import is_numeric_dtype
for i in df.columns:
  if(is_numeric_dtype(df[i])):
    mean = df[i].mean()
    df[i] = df[i].fillna(mean)
  else:
    df[i] = df[i].fillna("NA")

df.isna().sum()
```

```python
#Creating a class CustomLinearRegression to write linear regression from scratch
class CustomLinearRegression:
    def __init__(self):
        pass
    
    #The function fit to train our model on the dataset feeded to the model
    def fit(self, X, y, learning_rate=0.01, epochs=100):
        self.X = X
        self.Y = y
        self.learning_rate = learning_rate
        self.epochs = epochs
        #It is the theta as taught in class
        self.theta = np.zeros(X.shape[1],)
        #Storing the shape of the data
        self.m,self.n = X.shape
        #Calling the function gradient descent epoch times(which is the number of iterations) 
        self.gradient_descent()
        for i in range(self.epochs):
            self.gradient_descent()

    def loss(self):
        #making the prediction based on current theta values
        Y_pred = self.predict( self.X )
        #calculating the loss using MSE function (Mean Squared Error)
        dtheta = - ( ( self.X.T ).dot( self.Y - Y_pred )  ) / self.m
        return dtheta

    def gradient_descent(self):
        #using the gradient descent update rule taught in class
        #Updated the values of theta
        dtheta = self.loss()
        self.theta = self.theta - self.learning_rate * dtheta
        return self

    def predict(self, X):
        #returns the prediction by multiplying the matrices X and theta to get y (y = X.theta + error) here y,X,theta and error are matrices
        y_pred = X.dot(self.theta)
        return y_pred
```

```python
#Creating a class CustomGaussianNaiveBayesClassifier to implement it from scratch.
#Here I have implemented a custom approach that will work only on Iris dataset
class CustomGaussianNaiveBayesClassifier():  
  def __init__(self):
    pass

  #This function gets the X,y dataset and finds the probabilty of each class in the independent variable
  def separate_by_classes(self,X, y):
      #finding the frequency of each class
      self.classes, self.class_counts = np.unique(y, return_counts=True)
      class_freq = dict(zip(self.classes, self.class_counts))
      #finding probability of each class
      self.class_prob = {}
      for i in class_freq:
        self.class_prob[i] = class_freq[i]/len(y)
      #Plotting a bar graph between frequency and class using dict created 
      plt.bar(self.classes,self.class_counts)
      plt.xticks(range(3))
      plt.xlabel("Classes")
      plt.ylabel("No. of Entries")
      plt.title("Class Distribution")
      plt.show()

  #fit function to train our model
  def fit(self,X,y):
    self.separate_by_classes(X,y)
    #spliting the data frame into 3 parts according to its class
    #helps to find probability
    self.df = X.copy()
    self.df["Class"] = y
    self.df0 = self.df[self.df["Class"] == 0]
    self.df1 = self.df[self.df["Class"] == 1]
    self.df2 = self.df[self.df["Class"] == 2]
  
  #can predict for one row
  def predict_one(self,sl,sw,pl,pw):
    n0 = len(self.df0)
    n1 = len(self.df1)
    n2 = len(self.df2)
    
    #mean,std dev,number of sepal length
    msl = self.df["sepal length (cm)"].mean()
    msl0 = self.df0["sepal length (cm)"].mean() 
    msl1 = self.df1["sepal length (cm)"].mean() 
    msl2 = self.df2["sepal length (cm)"].mean() 

    ssl = self.df["sepal length (cm)"].std()
    ssl0 = self.df0["sepal length (cm)"].std() 
    ssl1 = self.df1["sepal length (cm)"].std() 
    ssl2 = self.df2["sepal length (cm)"].std() 

    # nsl = len(self.df[self.df["sepal length (cm)"] == sl])
    # nsl0 = len(self.df0[self.df0["sepal length (cm)"] == sl])
    # nsl1 = len(self.df1[self.df1["sepal length (cm)"] == sl])
    # nsl2 = len(self.df2[self.df2["sepal length (cm)"] == sl])

    #Calculating the normal probality
    #probability of given sepal length in dataset
    psl = self.calculate_probability(sl,msl,ssl)
    #probability of given sepal length given class 0
    psl0 = self.calculate_probability(sl,msl0,ssl0)
    #probability of given sepal length given class 1
    psl1 = self.calculate_probability(sl,msl1,ssl1)
    #probability of given sepal length given class 2
    psl2 = self.calculate_probability(sl,msl2,ssl2)

    #mean,std dev,number of sepal width
    msw = self.df["sepal width (cm)"].mean()
    msw0 = self.df0["sepal width (cm)"].mean() 
    msw1 = self.df1["sepal width (cm)"].mean() 
    msw2 = self.df2["sepal width (cm)"].mean() 

    ssw = self.df["sepal width (cm)"].std()
    ssw0 = self.df0["sepal width (cm)"].std() 
    ssw1 = self.df1["sepal width (cm)"].std() 
    ssw2 = self.df2["sepal width (cm)"].std() 

    # nsw = len(self.df[self.df["sepal width (cm)"] == sw])
    # nsw0 = len(self.df0[self.df0["sepal width (cm)"] == sw])
    # nsw1 = len(self.df1[self.df1["sepal width (cm)"] == sw])
    # nsw2 = len(self.df2[self.df2["sepal width (cm)"] == sw])

    #Calculating the normal probality
    #probability of given sepal width in dataset
    psw = self.calculate_probability(sw,msw,ssw)
    #probability of given sepal width given class 0
    psw0 = self.calculate_probability(sw,msw0,ssw0)
    #probability of given sepal width given class 1
    psw1 = self.calculate_probability(sw,msw1,ssw1)
    #probability of given sepal width given class 2
    psw2 = self.calculate_probability(sw,msw2,ssw2)
    
    #mean,std dev,number of petal length
    mpl = self.df["petal length (cm)"].mean()
    mpl0 = self.df0["petal length (cm)"].mean() 
    mpl1 = self.df1["petal length (cm)"].mean() 
    mpl2 = self.df2["petal length (cm)"].mean() 

    spl = self.df["petal length (cm)"].std()
    spl0 = self.df0["petal length (cm)"].std() 
    spl1 = self.df1["petal length (cm)"].std() 
    spl2 = self.df2["petal length (cm)"].std() 

    # npl = len(self.df[self.df["petal length (cm)"] == pl])
    # npl0 = len(self.df0[self.df0["petal length (cm)"] == pl])
    # npl1 = len(self.df1[self.df1["petal length (cm)"] == pl])
    # npl2 = len(self.df2[self.df2["petal length (cm)"] == pl])

    #Calculating the normal probality
    #probability of given petal length in dataset
    ppl = self.calculate_probability(pl,mpl,spl)    
    #probability of given petal length given class 0
    ppl0 = self.calculate_probability(pl,mpl0,spl0)
    #probability of given petal length given class 1
    ppl1 = self.calculate_probability(pl,mpl1,spl1)
    #probability of given petal length given class 2
    ppl2 = self.calculate_probability(pl,mpl2,spl2)

    #mean,std dev,number of petal width
    mpw = self.df["petal width (cm)"].mean()
    mpw0 = self.df0["petal width (cm)"].mean() 
    mpw1 = self.df1["petal width (cm)"].mean() 
    mpw2 = self.df2["petal width (cm)"].mean() 

    spw = self.df["petal width (cm)"].std()
    spw0 = self.df0["petal width (cm)"].std() 
    spw1 = self.df1["petal width (cm)"].std() 
    spw2 = self.df2["petal width (cm)"].std() 

    # npw = len(self.df[self.df["petal width (cm)"] == pw])
    # npw0 = len(self.df0[self.df0["petal width (cm)"] == pw])
    # npw1 = len(self.df1[self.df1["petal width (cm)"] == pw])
    # npw2 = len(self.df2[self.df2["petal width (cm)"] == pw])

    #Calculating the normal probality
    #probability of given petal width in dataset
    ppw = self.calculate_probability(pw,mpw,spw)
    #probability of given petal width given class 0
    ppw0 = self.calculate_probability(pw,mpw0,spw0)
    #probability of given petal width given class 1
    ppw1 = self.calculate_probability(pw,mpw1,spw1)
    #probability of given petal width given class 2
    ppw2 = self.calculate_probability(pw,mpw2,spw2)

    #calculating the probability of class 0 given the conditions using the bayes theorem formula
    p0 = (psl0*psw0*ppl0*ppw0*self.class_prob[0])/(psl*psw*ppl*ppw)

    #calculating the probability of class 1 given the conditions using the bayes theorem formula
    p1 = (psl1*psw1*ppl1*ppw1*self.class_prob[1])/(psl*psw*ppl*ppw)

    #calculating the probability of class 2 given the conditions using the bayes theorem formula
    p2 = (psl2*psw2*ppl2*ppw2*self.class_prob[2])/(psl*psw*ppl*ppw)

    #The class with maximum probability will be predicted according to the algorithmn 
    if(p0 >= p1 and p0>=p2):
      return 0
    elif(p1 >= p0 and p1>=p2):
      return 1
    elif(p2 >= p0 and p2>=p1):
      return 2  

  def calculate_probability(self,x, mean, stddev):
    #Calculate the normal probability of given point x w.r.t. mean and stddev
    exponent = math.exp(-(math.pow((x - mean),2) / (2 * stddev ** 2)))
    return (1 / (math.sqrt(2 * math.pi) * stddev)) * exponent

  def predict(self,X_test):
    #Predict the by looping over X_test and getting predictions one at a time from predict_one function
    rows = X_test.shape[0]
    y_predict = np.zeros(rows,)
    for i in range(rows):
      x = X_test.iloc[i]
      y_predict[i] = self.predict_one(x[0],x[1],x[2],x[3])
    return y_predict
```

```python
#Creating a class CustomKNearestNeighbour to implement our custom KNN
class CustomKNearestNeighbour:
  def __init__(self,k):
    self.k = k

  # fitting the model on given dataset
  def fit(self,X_train,y_train):
    self.X_train = X_train
    self.y_train = y_train
  
  #given a point find its dist from all the points in X_train and returns a sorted list
  def calculate_dist(self,point):
    dist_dict = {}
    #looping and finding euc dist
    for i in range(len(self.X_train)):
      euc_dist = abs(X_train[i] - point) 
      dist_dict[i] = euc_dist
    #sorted dict of distances
    sorted_dict = sorted(dist_dict.items(), key=lambda x:x[1])
    return sorted_dict

  #can predict for one row
  def predict_one(self,point):
    sorted_dict = self.calculate_dist(point)
    total = 0
    # get first k nearest neighbours
    for i in range(self.k):
      total += self.y_train[sorted_dict[i][0]]
    #returns average of the neighbours
    return total/k

  
  def predict(self,X_test):
    #Predict the by looping over X_test and getting predictions one at a time from predict_one function
    rows = X_test.shape[0]
    y_pred = np.zeros(rows,)
    for i in range(rows):
      y_pred[i] = self.predict_one(X_test[i])
    return y_pred

  #To find R^2 error between y_pred and y_test
  def R_Square_Error(self,y_pred,y_test):
    #simply using the formula taugt in class
    y_mean = y_test.mean()
    SSR = 0
    SST = 0
    for i in range(len(y_test)):
      SSR += ((y_pred[i] - y_test[i])*(y_pred[i] - y_test[i]))
      SST += ((y_test[i] - y_mean)*(y_test[i] - y_mean)) 
    return float(1 - (SSR/SST))
```

```python
# for k values in 1 to 16 I have printed its R^2 error
r2_errors = []
for k in range(1,16):
  # using my custom class created a model
  model = CustomKNearestNeighbour(k)
  #fitting the model
  model.fit(X_train,y_train)
  #predicting from the model
  y_pred = model.predict(X_test)
  #printing the error for each value of k
  error = model.R_Square_Error(y_pred,y_test)
  r2_errors.append(error)
  print("R^2 Error for K = ",k," is ",error)

```
